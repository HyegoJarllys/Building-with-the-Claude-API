```python
from dotenv import load_dotenv
load_dotenv()
from anthropic import Anthropic
client = Anthropic()
model="claude-sonnet-4-0"

system_prompt='''
You Are a Assistent in python sênior with 50 years of carrer
your function is datail everyone code and fundaments, step by step and clean didatic'''
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1024,
    system=system_prompt,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

message.content[0].text
```

```python
#funções que ajudam na construção de mensagens e contexto para o chat
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=0.2):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature":temperature
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# começa com uma lista vazia de mensagens
messages = []

# adiciona a primeira pergunta do usuário
add_user_message(messages, "Define quantum computing in one sentence")
with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        # enviar cada token por vez
        pass
    
    # pegar a mensagem final para a funcao
    final_message = stream.get_final_message()

answer = chat(messages)

# adiciona a resposta do assistente à história da conversa
add_assistant_message(messages, answer)

# adiciona uma pergunta de follow-up do usuário
add_user_message(messages, "Write another sentence")

# adiciona a resposta do assistente à história da conversa
final_answer = chat(messages)
```
